```python
data = data[['created_at', 'PM1.0_CF1_ug/m3', 'PM2.5_CF1_ug/m3', 'PM10.0_CF1_ug/m3',
       'UptimeMinutes', 'RSSI_dbm', 'Temperature_F', 'Humidity_%',
       'PM2.5_ATM_ug/m3']].copy()
```

```python
data[['year','month','day']] = data.Time.str.split('-',expand = True)
```

```python
data[['year','month','day']]
```

```python
#QUICKLOOK
data.groupby('month')['PM2.5_CF1_ug/m3'].mean().plot()
```

```python
monthly_pm = data.groupby('month')['PM2.5_CF1_ug/m3'].mean()
monthly_temp = data.groupby('month')['Temperature_F'].mean()
monthly_hum = data.groupby('month')['Humidity_%'].mean()
```

```python
monthly_temp.plot()
```

```python
monthly_hum.plot()
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug']

'''

fig = plt.figure(figsize=(10, 5))

plt.bar(data.month.unique(), monthly_pm, color='red', label='PM2.5')
plt.bar(data.month.unique(), monthly_temp, color='orange', label='Temperature')
plt.bar(data.month.unique(), monthly_hum, color='blue', label='Humidity')

plt.xticks(ticks=data.month.unique(), labels=months)
plt.xlabel('Months')
plt.ylabel('Values')

plt.grid(True)  
plt.title('Monthly Variation of PM2.5, Temperature, and Humidity')
plt.legend()  

plt.tight_layout()
plt.show()
'''
```

```python
fig = plt.figure(figsize=(15, 7))

# Generate an array for x-coordinates 
x = np.arange(len(data.month.unique()))

plt.bar(x - 0.2, monthly_pm, color=plt.cm.terrain(0.4), width=0.2, label='PM2.5')
plt.bar(x, monthly_temp, color=plt.cm.terrain(0.6), width=0.2, label='Temperature')
plt.bar(x + 0.2, monthly_hum, color=plt.cm.terrain(0.8), width=0.2, label='Humidity')

plt.xticks(ticks=x, labels=months)
plt.xlabel('Months')
plt.ylabel('Values')

#plt.grid(True)  
plt.title('Monthly Variation of PM2.5, Temperature, and Humidity')
plt.legend()  

plt.tight_layout()
plt.savefig('visuals/MONTHLY VARIATION OF PM2.5.png',dpi=300,facecolor='w', pad_inches=0.1)
```

```python
fig,ax=plt.subplots(1,3,figsize=(9,5))

fig.tight_layout()


ax[0].plot(data.month.unique(),monthly_pm,marker='o')
ax[0].set_xticks(data.month.unique().tolist(),labels = months)
ax[0].set_xlabel('MONTHS')
ax[0].set_ylabel('PM2.5')

ax[1].plot(data.month.unique(),monthly_temp,marker='o')
ax[1].set_xticks(data.month.unique().tolist(),labels = months)
ax[1].set_xlabel('MONTHS')
ax[1].set_ylabel('TEMPERATURE')

ax[2].plot(data.month.unique(),monthly_hum,marker = 'o')
ax[2].set_xticks(data.month.unique().tolist(),labels = months)
ax[2].set_xlabel('MONTHS')
ax[2].set_ylabel('HUMIDITY');

```

```python
fig, ax = plt.subplots(1, 3, figsize=(15, 5))  

# Plot PM2.5
ax[0].plot(data.month.unique(), monthly_pm, marker='o', color='b', label='PM2.5')
ax[0].set_xticks(data.month.unique().tolist())
ax[0].set_xticklabels(months, rotation=45)  
ax[0].set_xlabel('Months')
ax[0].set_ylabel('PM2.5')
ax[0].grid(True)  
ax[0].set_title('Monthly PM2.5 Variation')  
ax[0].legend()

# Plot Temperature
ax[1].plot(data.month.unique(), monthly_temp, marker='s', color='r', label='Temperature')
ax[1].set_xticks(data.month.unique().tolist())
ax[1].set_xticklabels(months, rotation=45)
ax[1].set_xlabel('Months')
ax[1].set_ylabel('Temperature')
ax[1].grid(True)
ax[1].set_title('Monthly Temperature Variation')
ax[1].legend()

# Plot Humidity
ax[2].plot(data.month.unique(), monthly_hum, marker='^', color='g', label='Humidity')
ax[2].set_xticks(data.month.unique().tolist())
ax[2].set_xticklabels(months, rotation=45)
ax[2].set_xlabel('Months')
ax[2].set_ylabel('Humidity')
ax[2].grid(True)
ax[2].set_title('Monthly Humidity Variation')
ax[2].legend()
plt.suptitle('Relationship Between Pm2.5, Temperature and Humidity')

plt.tight_layout()
```
