```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_excel('1.xlsx')
data
```

```python
data.notnull
```

```python
#quicklook
data.groupby('time')['PM2.5_CF1_ug/m3'].mean().plot()
```

```python
hourly_pm = data.groupby('time')['PM2.5_CF1_ug/m3'].mean()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(data.time.unique(),hourly_pm,color='black')
plt.xlabel('TIME')
plt.ylabel('PM2.5_CF1_ug/m3')
plt.title('HOURLY DISTRIBUTION OF PM2.5')
plt.xticks(rotation=45);

plt.savefig('visuals/HOURLY DISTRIBUTION OF PM2.5.png',dpi = 300,facecolor='w', pad_inches=0.1)
```

```python
hourly_pm10=data.groupby('time')['PM10.0_CF1_ug/m3'].mean()
```

```python
fig = plt.figure(figsize=(10,5))
plt.plot(data.time.unique(),hourly_pm10,color='black')
plt.xlabel('TIME')
plt.ylabel('PM10.0_CF1_ug/m3')
plt.title('HOURLY DISTRIBUTION OF PM10')
plt.xticks(rotation = 45);

plt.savefig('visuals/HOURLY DISTRIBUTION OF PM10.png',dpi = 300,facecolor='w', pad_inches=0.1)
```
